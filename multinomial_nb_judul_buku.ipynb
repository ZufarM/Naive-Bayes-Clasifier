```python
from sklearn.feature_extraction.text import CountVectorizer
data_1 = [df['judul'][9]]

# create the transform
vectorizer = CountVectorizer()

vectorizer.fit(data_1)


print(vectorizer.vocabulary_) 
```

```python
import string
from nltk.corpus import stopwords
def text_cleaning(a):
 remove_punctuation = [char for char in str(a) if char not in string.punctuation]
 #print(remove_punctuation)
 remove_punctuation=''.join(remove_punctuation)
 #print(remove_punctuation)   
 return [word for word in remove_punctuation.split() if word.lower() not in stopwords.words('indonesian')]
```

```python
print(df.iloc[:,2].apply(text_cleaning))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=text_cleaning).fit(df['judul_tokens_stemmed'].values.astype('U')) 
  
#print(len(bow_transformer.vocabulary_))   
bow_transformer.vocabulary_
```

```python
title_bow = bow_transformer.transform(df['judul_tokens_stemmed'].values.astype('U'))

print(title_bow)
```

```python
X = title_bow.toarray()
print(X)
 
X.shape 
```

```python
# TF-IDF Algo -term frequency-inverse document frequency to know the most significant words

from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer=TfidfTransformer().fit(title_bow)
print(tfidf_transformer)

title_tfidf=tfidf_transformer.transform(title_bow)
print(title_tfidf)# got tfidf values for whole vocabulary
print(title_tfidf.shape)  
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(
    title_tfidf, # data training 80%
    df['no']
)
```

```python
all_predictions = model.predict(
    title_tfidf # data uji 20%
)
print(all_predictions)
```

```python
#Printing the confusion matrix of our prediction
from sklearn.metrics import confusion_matrix

confusion_matrix(
    df['no'], 
    all_predictions
)
```
