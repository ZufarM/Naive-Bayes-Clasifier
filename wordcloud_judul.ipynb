```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load in the dataframe
df = pd.read_csv("data/Text_Preprocessing.csv")
df.head()
```

```python
class3 = df.loc[df.no == 3]
class3.judul_tokens_stemmed
```

```python
wordcloud = WordCloud().generate(class3.judul_tokens_stemmed[0])

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
text = df.loc[df.no == 3]

text = " ".join(review for review in map(str,text.judul_tokens_stemmed))
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["',", "now", "wine", "flavor", "flavors"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
def cek_wordcloud_by_class_function(x):
    text = df.loc[df.no == x]
    text = " ".join(review for review in map(str,text.judul_tokens_stemmed))
    print ("Kode Kelas {}".format(str(x)))
    print ("There are {} words in the combination of all review.".format(len(text)))

    
    # Create stopword list:
    stopwords = set(STOPWORDS)
    stopwords.update(["',", "now", "wine", "flavor", "flavors"])

    # Generate a word cloud image
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

    # Display the generated image:
    # the matplotlib way:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
